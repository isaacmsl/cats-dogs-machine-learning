```python
import pandas as pd
from skimage.feature import hog
from keras.preprocessing.image import load_img
import os


resize_tuple = (128, 128)
pixels_per_cel = (20, 20)
cells_per_block = (2, 2)
```

```python
image_paths = []
for filename in os.listdir('../images'):
    if filename.endswith('.jpg'):
        image_paths.append(os.path.join('../images', filename))

image_paths
```

```python
hog_features = []
for image_path in image_paths:
    print(f"Extracting features of {image_path}...")
    image_resized = load_img(image_path, target_size=resize_tuple)
    fd, hog_image = hog(image_resized, orientations=9, pixels_per_cell=pixels_per_cel,
                        cells_per_block=cells_per_block, visualize=True, channel_axis=-1)
    hog_features.append(fd)
```

```python
classe = [
    (
        "american_bulldog"
        if "american_bulldog" in file
        else (
            "american_pit_bull_terrier"
            if "american_pit_bull_terrier" in file
            else (
                "bengal"
                if "Bengal" in file
                else "abyssinian" if "Abyssinian" in file else file
            )
        )
    )
    for file in image_paths
]
```

```python
data = {"classe": classe, "hog_features": hog_features}
df = pd.DataFrame(data)
df.head()
```

```python
# transformando cada valor do hog_feature em um atributo (coluna)
df2 = pd.DataFrame(df['hog_features'].tolist())
df2.columns = df2.columns.map(lambda x: f'hog_feature_{x+1}')
df = pd.concat([df.drop('hog_features', axis=1), df2], axis=1)


df.info()
```
